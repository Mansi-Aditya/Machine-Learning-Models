```python
import tensorflow as tf             #importing tensorflow library
import numpy as np                  #importing numpy library
from tensorflow import keras         # importing keras in tensorflow
#It is used to tell that we use single dense layer . where units = 1 is for a single neuron prsent in layer
# and input size is for input value will be a single value
model= tf.keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
# This line configures the model with Stochastic Gradient Descent as the optimizer and Mean Squared Error as the loss function.
model.compile(optimizer='sgd', loss='mean_squared_error')
#giving the sample data using numpy
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
#Train the neural network
model.fit(xs, ys, epochs=500)
#Predicting the model
# Convert the input to a NumPy array
print(model.predict(np.array([10.0])))
```
